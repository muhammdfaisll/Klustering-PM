**MEMBUAT NILAI X**

```python
x_array = np.array(data)
```

**VISUALISASI PERSEBARAN DATA**

```python
plt.scatter(data.W1, data.W2)
plt.show()
```

**STANDARISASI**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_array)
plt.scatter(x_scaled[:,0], x_scaled[:,1], s=10)
plt.show()
```

**CEK SUM OF SQUARE ERROR DARI TIAP PEMBAGIAN JUMLAH CLUSTER**

```python
from sklearn.cluster import KMeans
sse = []
index = range(1,10)
for i in index :
  kmeans = KMeans(n_clusters = i, random_state = 0, n_init = 'auto')
  kmeans.fit(x_scaled)
  sse_ = kmeans.inertia_
  sse.append(sse_)
  print(i,sse)
```

```python
plt.plot(index, sse)
plt.show()
```

**MEMBUAT MODEL**

```python
kmeans = KMeans (n_clusters = 3, random_state = 0)
kmeans.fit(x_scaled)
```

**MELIHAT CLUSTER PUSAT**

```python
kmeans.cluster_centers_
```

**VISUALISAI PESABARAN DATA SETELAH CLUSTERING**

```python
output = plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=kmeans.labels_)

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c="red", s=10)  # Fix the indexing here
plt.title("KMEANS Clustering Credit Card")
plt.xlabel("W1")
plt.ylabel("W2")
plt.colorbar(output)
plt.show()
```

**EVALUASI MODEL**

```python
from sklearn.metrics import davies_bouldin_score
labels = kmeans.labels_
davies_bouldin_score(x_scaled, labels)
```
